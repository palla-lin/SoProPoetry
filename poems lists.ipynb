### Prepocessing

- create lists with poems: lowercase and keep punctuation as separate tokens
- remove poems without topics (tags)
- remove poems that are not broken down to stanzas (length line is over a threshold) **?** 


```python
import pandas as pd
# Read csv file into a pandas dataframe
csv_path = 'PoetryFoundationData.csv'
df = pd.read_csv(csv_path)
display(df)
```

```python
# remove poems that don't have tags/topics

df.dropna(subset=['Tags'],inplace=True)
df.reset_index(drop=True, inplace=True)
display(df)
# 12899 poems after removing poems without tags
```

```python
df['Tags'].isnull().sum()
```

```python
import re

def preprocessing(text): 
    poem_words = []
    for line in text.split("\r\r\n"):
        if len(line) > 150: # remove poems without stanzas
            poem_words = [] 
            break
        res = re.findall(r"[\w']+|[.,!?;:]", line.lower())
        poem_words += res
    return poem_words    
    
# Preprocess poems 

poems = df['Poem']
# display(poems)
poems_list = []
c=[]
for i in range(len(poems)):
    poem = poems[i]
    if preprocessing(poem) != []:
        poems_list.append(preprocessing(poem))
        
# print(poems_list[5:10])
```
